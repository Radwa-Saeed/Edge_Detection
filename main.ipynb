# Hough Lines

## Example 1

### 1- Loading the original Image 

```python
path1 = "./images/building.png"

line_image1 = Read_Img(path1)
line_image1 = cv2.cvtColor(line_image1, cv2.COLOR_BGR2RGB)

Display_Img("Original",line_image1)
```

### 2-  Applying Hough Line

```python
hough_lines_output1 = hough_line(img = line_image1, min_line_votes = 225)

Display_Img("Hough Line ", hough_lines_output1)
```

### 3-  Superimposing the Hough Line output binary image over the Original image

```python
hough_lines_output1_sup = edgeSuperimpose(image = line_image1, edges = hough_lines_output1, color = "green")

Display_Img("Hough Line Superimposed", hough_lines_output1_sup)
```

# Hough Circles

## Example 1


### 1- Creating Image that contain different circles

```python
circle_image = 255*np.zeros((256,256,3), dtype=np.uint8)

circles = [(50,50,25),(150,150,20),(200,200,20),(100,100,30),(50,200,20)]

for x , y , r in circles:  
    circle_image = cv2.circle(circle_image, (x, y),r, (255,0,0), -1,lineType=cv2.LINE_AA)

Display_Img ("Original",circle_image)
```

### 2- Applying Hough Circles

```python
hough_circles_output1  = hough_circles(circle_image, n_circles = 5, minRadius = 15, maxRadius = 35)

Display_Img ("Hough Circle", hough_circles_output1)
```

### 3-  Superimposing the Hough Circles output binary image over the Original image

```python
hough_circles_output1_sup = edgeSuperimpose(image = circle_image, edges = hough_circles_output1, color = "green")

Display_Img("Hough Circle Superimposed", hough_circles_output1_sup)
```
